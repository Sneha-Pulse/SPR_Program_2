#  Applying Fourier Transform on Various Signals for Frequency Analysis  







```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftshift
from scipy.signal import freqz
```

# PART 1








## (A) Generate a basic sinusoidal signal in the time domain. (For example, you may generate a sine wave with a frequency of 5Hz, sampled at 1000Hz.)

```python
# (a) Generate a sinusoidal signal in the time domain
fs = 1000  # Sampling frequency (Hz)
f = 5      # Frequency of the sine wave (Hz)
T = 1      # Duration (seconds)
t = np.linspace(0, T, fs, endpoint=False)  # Time vector
x = np.sin(2 * np.pi * f * t)  # Sine wave
```

## (B) Plot the time-domain waveform of the signal.

```python
# (b) Plot the time-domain waveform
plt.figure(figsize=(10, 4))
plt.plot(t, x)
plt.title('Time-Domain Signal: 5 Hz Sine Wave')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## (B) Compute the Discrete-Time Fourier Transform (DTFT) and plot the continuous frequency spectrum.

```python
# (c) Compute the Discrete-Time Fourier Transform (DTFT)
# Using the frequency response function freqz
frequencies, X_dtft = freqz(x, 1, worN=8000, fs=fs)
```

```python
# Plot the DTFT (continuous frequency spectrum)
plt.figure(figsize=(10, 4))
plt.plot(frequencies, np.abs(X_dtft))
plt.title('Continuous Frequency Spectrum (DTFT)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

### (D) Compute the Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum.

 The Discrete Fourier Transform (DFT) is a mathematical transformation used to convert a sequence of N complex numbers (a discrete-time signal) into another sequence of complex numbers, which represents the signal in the frequency domain. The DFT is essentially a sampled version of the Continuous Fourier Transform, and it is used extensively in digital signal processing and related fields for analyzing the frequency content of discrete-time signals.


Mathematically, if x[n] represents the discrete-time signal, the DFT X[k] of length N is defined as:
image.png


In this equation, X[k] represents the k-th frequency component in the frequency domain, while x[n] represents the time-domain signal. The DFT provides information about the amplitudes and phases of various frequency components present in the signal.


The DFT is computed using algorithms like the Fast Fourier Transform (FFT), which significantly speeds up the calculation process for practical applications. It has wide-ranging applications in fields such as audio signal processing, image processing, telecommunications, and many more, making it a fundamental concept in the study and application of digital signal processing.  

```python
# (d) Compute the Discrete Fourier Transform (DFT)
X_dft = fft(x)  # DFT of the signal
N = len(X_dft)  # Number of points in the DFT
frequencies_dft = np.fft.fftfreq(N, 1/fs)  # Frequency bins
```

```python
# Plot the DFT (discrete frequency spectrum)
plt.figure(figsize=(10, 4))
plt.stem(frequencies_dft[:N//2], np.abs(X_dft)[:N//2], basefmt=" ")
plt.title('Discrete Frequency Spectrum (DFT)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
from scipy.signal import freqz
```

## (A) Generate a composite signal by adding two or more sinusoidal signals of different frequencies and amplitudes.

```python
# (a) Generate a composite signal by adding two sinusoidal signals
fs = 1000  # Sampling frequency (Hz)
T = 1      # Duration (seconds)
t = np.linspace(0, T, fs, endpoint=False)  # Time vector

# Sinusoidal signal 1: frequency = 5Hz, amplitude = 1
f1 = 5
A1 = 1
x1 = A1 * np.sin(2 * np.pi * f1 * t)

# Sinusoidal signal 2: frequency = 50Hz, amplitude = 0.5
f2 = 50
A2 = 0.5
x2 = A2 * np.sin(2 * np.pi * f2 * t)

# Add the two signals to create the composite signal
composite_signal = x1 + x2
```

## (B) Plot the time-domain waveform of the composite signal.

```python
# (b) Plot the time-domain waveform of the composite signal
plt.figure(figsize=(10, 4))
plt.plot(t, composite_signal)
plt.title('Time-Domain Signal: Composite Signal (5Hz and 50Hz)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## (C) Compute the Discrete-Time Fourier Transform (DTFT) and plot the continuous frequency spectrum.

```python
# (c) Compute the Discrete-Time Fourier Transform (DTFT)
frequencies, X_dtft = freqz(composite_signal, 1, worN=8000, fs=fs)
```

```python
# Plot the DTFT (continuous frequency spectrum)
plt.figure(figsize=(10, 4))
plt.plot(frequencies, np.abs(X_dtft))
plt.title('Continuous Frequency Spectrum (DTFT) of Composite Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```










## (D) Compute the Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum.

```python
# (d) Compute the Discrete Fourier Transform (DFT)
X_dft = fft(composite_signal)  # DFT of the composite signal
N = len(X_dft)  # Number of points in the DFT
frequencies_dft = np.fft.fftfreq(N, 1/fs)  # Frequency bins
```

```python
# Plot the DFT (discrete frequency spectrum)
plt.figure(figsize=(10, 4))
plt.stem(frequencies_dft[:N//2], np.abs(X_dft)[:N//2], basefmt=" ")
plt.title('Discrete Frequency Spectrum (DFT) of Composite Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

## (A) Generate an exponentially decaying signal.

###  An exponentially decaying signal:

 An exponentially decaying signal is a type of signal whose magnitude decreases exponentially over time.

 Mathematically, an exponentially decaying signal can be represented as

>  s(t) = A * exp(-αt)





 where A is the initial amplitude, t is time, and α is the decay constant. These signals are commonly encountered in various fields such as physics, engineering, and biology, and they are used to model phenomena such as radioactive decay, capacitor discharging, and damped oscillations.  

```python
# (a) Generate an exponentially decaying signal
fs = 1000  # Sampling frequency (Hz)
T = 1      # Duration (seconds)
t = np.linspace(0, T, fs, endpoint=False)  # Time vector
tau = 0.1  # Time constant of the decay

# Exponentially decaying signal: A * exp(-t/tau)
A = 1  # Amplitude of the signal
decaying_signal = A * np.exp(-t / tau) * np.cos(2 * np.pi * 5 * t)  # Cosine modulated by exponential decay
```

## (B) Plot the time-domain waveform.

```python
# (b) Plot the time-domain waveform
plt.figure(figsize=(10, 4))
plt.plot(t, decaying_signal)
plt.title('Time-Domain Signal: Exponentially Decaying Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## (C) Compute the Discrete-Time Fourier Transform (DTFT) and plot the continuous frequency spectrum.

```python
# (c) Compute the Discrete-Time Fourier Transform (DTFT)
frequencies, X_dtft = freqz(decaying_signal, 1, worN=8000, fs=fs)
```

```python
# Plot the DTFT (continuous frequency spectrum)
plt.figure(figsize=(10, 4))
plt.plot(frequencies, np.abs(X_dtft))
plt.title('Continuous Frequency Spectrum (DTFT) of Exponentially Decaying Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

## (D) Compute the Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum.

```python
# (d) Compute the Discrete Fourier Transform (DFT)
X_dft = fft(decaying_signal)  # DFT of the signal
N = len(X_dft)  # Number of points in the DFT
frequencies_dft = np.fft.fftfreq(N, 1/fs)  # Frequency bins
```

```python
# Plot the DFT (discrete frequency spectrum)
plt.figure(figsize=(10, 4))
plt.stem(frequencies_dft[:N//2], np.abs(X_dft)[:N//2], basefmt=" ")
plt.title('Discrete Frequency Spectrum (DFT) of Exponentially Decaying Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```
